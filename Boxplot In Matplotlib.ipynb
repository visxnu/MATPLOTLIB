```python
import matplotlib.pyplot as plt
import numpy as np

data=[20,15,18,22,24,19,30,25,34,28,25]

plt.boxplot(data)

plt.title("boxplot exmple")
plt.xlabel("Data category")
plt.ylabel("values")

plt.show()
```

```python
x=[55,60,62,65,67,70,72,75,50,85,90,95,98]
y=[45,50,52,55,60,62,65,68,70,72,75,78,80]

plt.figure(figsize=(8,6))
plt.boxplot([x,y],labels=['Group A','Group B'])

plt.title('class score distribution')
plt.xlabel("classes")
plt.ylabel("scores")

plt.show()
```

```python
x=[55,60,62,65,67,70,72,75,50,85,90,95,98]
y=[45,50,52,55,60,62,65,68,70,72,75,78,80]

plt.figure(figsize=(8,6))
plt.boxplot([x,y],labels=['Group A','Group B'],patch_artist=True)

plt.title('class score distribution')
plt.xlabel("classes")
plt.ylabel("scores")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [20, 15, 18, 22, 24, 19, 30, 32, 45, 28, 25]

plt. boxplot(data, patch_artist=True, boxprops=dict(facecolor= 'lightblue', color= 'blue'),
             whiskerprops=dict(color='blue'),capprops=dict(color='red'),
             flierprops=dict (markerfacecolor='green', marker='o', markersize=7))

plt.title( 'Boxplot Example')
plt.xlabel ('Data Category')
plt.ylabel( 'Values')

plt.show()
```

```python
import matplotlib.pyplot as plt


data1 = [20, 15, 18, 22, 24, 19, 30, 32, 34, 28, 25]
data2 = [10, 12, 14, 18, 19, 21, 24, 25, 27, 29, 31]

fig, axs = plt. subplots(1, 2, figsize=(12, 6))

axs[0].boxplot (data1,patch_artist=True)

axs[0].set_title('Boxplot 1')
axs[0].set_xlabel('Category')
axs[0].set_ylabel('Values')

axs[1].boxplot (data2,patch_artist=True)
axs[1].set_title('Boxplot 2')
axs[1].set_xlabel('Category')
axs[1].set_ylabel('Values')

plt. tight_layout ()
plt. show()
```
