# 3D plots

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=plt.axes(projection='3d')
```

# 3D Line Chart

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot(x,y,z,color='purple')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Simple 3D Line Chart')

plt.show()
```

# Annotation

```python
import matplotlib.pyplot as plt
# sample data

x = [1,2,3,4,5]
y = [10,15,13,18,16]

#Plot the data
plt.plot(x,y,marker='o')

#Add an annotation
plt.annotate('Highest Point',xy=(4,18),xytext=(3,20),
             arrowprops=dict(arrowstyle='->',color='red'))

#labels
plt.xlabel('X Axis')
```

```python
x = [1,2,3,4,5]
y = [10,15,13,17,16]


plt.plot(x,y,marker='o',color='b',linestyle='-',label='Sample Line')
plt.annotate('Highest Point',xy=(4,17),xytext=(3.5,18),arrowprops=dict(arrowstyle='->',color='r'))


for i,j in zip(x,y):
  plt.text(i,j+0.5,f'({i})',ha='center',color='green')


plt.legend()

plt.show()
```

```python
x=[1,2,3,4,5]
y=10,15,13,17,16

plt.plot(x,y,marker='o',color='b',linestyle='-',label='Sample Line')

for i,j in zip(x,y):
  plt.text(i,j,f'({i} , {j})',ha='center',color='green')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

categories=['A','B','C','D']
values=[10,15,7,12]

plt.bar(categories,values,color="skyblue")

for i,j in enumerate(values):
  plt.text(i,j+0.3,str(j),ha='center',color='black')

plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart with annotations')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,15,13,18,16]

plt.scatter(x,y,color="green")

for i,j in zip(x,y):
  plt.annotate(f'({i},{j})',xy=(i,j),xytext=(i,j+0.5),color="blue")

plt.xlabel("X_axis")
plt.ylabel('y_label')
plt.title('sactter plot with annotation ')
plt.show()
```
