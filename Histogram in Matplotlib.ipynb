**A histogram is a graph showing Frequency distribustions**

kde =Kernel Distribustion Estimation

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10 ,250)

plt.hist(x)
plt.show()
```

# Bins

Bins are used to define count of histogram

```python
#bins = None
#       sequnce
#       int

data = [1,12,6,21,22,20,21,24]

plt.hist(data,4,ec = 'red')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
data = [1,12,6,21,22,20,21,24]

plt.hist(data,bins=[1,5,10,20,30],ec = 'red')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

# Range

```python
data = [1,15,10,16,5,7,25,6]

plt.hist(data,range=(0,25))
plt.show()
```

# Density

```python
import matplotlib.pyplot as plt

data = [1,12,6,21,22,20,21,24]

plt.hist(data, 3, density=True, ec='red')


plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1,12,6,21,22,20,21,24]

plt.hist(data, 3, density=False, ec='red')


plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```
