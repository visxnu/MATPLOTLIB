# Createing Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["One","Two","Three","Four","Five"])
y = np.array([5,25,35,68,10])

plt.title("Vertical bar graph")
plt.xlabel("Pen sold")
plt.ylabel("Price")
plt.bar(x,y)
plt.show()
```

# Bar Width

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.array([4,9,2,2,9,4])
y = np.arange(6)

plt.barh(y,x,color = 'r')
plt.barh(y,-x,color = 'black')


plt.legend(["A","B"])
plt.xlabel("Data")
plt.ylabel("Price")

plt.title('Back to Backbar')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#set width of bar

barWidth = 0.25
fig = plt.subplots(figsize=(10,4))

# set height of bar

IT = [12,30,1,8,22]
ECE = [28,6,16,5,10]
CSE = [29,3,24,25,17]

#Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
plt.bar(br1, IT, color = 'r',width = barWidth,edgecolor = 'grey', label='IT')
plt.bar(br2, ECE, color = 'g',width = barWidth,edgecolor = 'grey', label='ECE')
plt.bar(br3, CSE, color = 'b',width = barWidth,edgecolor = 'grey', label='CSE')

plt.xlabel("Branch",fontweight ='bold',fontsize=15)
plt.ylabel("Students Passed",fontweight = 'bold',fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],['2015','2016','2017','2018','2019'])
plt.legend()
plt.show()
```

```python
N = 5

boys = (20,35,30,35,27)
girls = (25,32,34,20,25)
boyStd = (2,3,4,1,2)
girlStd = (3,5,2,3,3)
ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize = (10,7))
p1 = plt.bar(ind,boys,width,yerr = boyStd)
p2 = plt.bar(ind,girls,width,bottom=boys,yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('boys','girls'))

plt.show()
```
