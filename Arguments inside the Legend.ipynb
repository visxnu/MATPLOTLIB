# Prop Argument

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5))
plt.plot([1,2,3,6,5])
plt.plot([3,5,4,8,6])
plt.plot([1,4,6,3,5])
plt.plot([7,2,1,5,4])


plt.legend(['blue','orange','green','red'],prop={'size' :10},loc="upper left",shadow=True,facecolor='yellow',labelspacing=1)
plt.show()
```

# Label spaceing

```python
x = [1,2,3,4,5]
y = [3,3,3,3,3]

plt.plot(x,y,label='line1')
plt.plot(y,x,label='line2')
plt.plot(x,np.sin(x),label='line3')
plt.plot(x,np.cos(x),label='line4')

plt.legend(labelspacing=1)
plt.title("Hai Heloo")
plt.show()
```

# N col

# Patches and handles

```python
import matplotlib.patches as mpatches

plt.plot([1,2,3,4],color = 'blue')
plt.title("Simple Legend")
blue_patch = mpatches.Patch(color="blue",label="Blue Color")
plt.legend(handles=[blue_patch])
plt.show()
```

# Tight Layout

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),label="sin x")
plt.plot(x,np.cos(x),label="cos x")

plt.legend(bbox_to_anchor=(1.05,1.0),loc='upper left')
plt.tight_layout()

plt.show()
```

# Frame on

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,10,100)

y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['Sine wave','Cos wave'],frameon=False)
plt.show()
```

# Annotenation

```python
import matplotlib.pyplot as plt
categoies = ['A','B','C']
values = [10,20,15]

plt.bar(categoies,values,color='skyblue')

for i,v in enumerate(values):
  plt.text(i,v,str(v))

plt.xlabel('Categories')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

# Stacked

```python
import numpy as np

a_ages = [22,25,29,30,34,36,28]
b_ages = [24,26,30,32,35,38,27]

bins = np.arange(20,40,5)

plt.hist(a_ages,bins=bins,ec='k',alpha=0.7,label='Group A',stacked=True)
plt.hist(b_ages,bins=bins,ec='k',alpha=0.7,label='Group B',stacked=True)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Stacked Histogram of Age Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60,15,100)

bins = np.linspace(20,100,20)

plt.figure(figsize=(10,10))

plt.hist(data1,bins=bins,color='blue',alpha=0.6,label='Data 1')
plt.hist(data2,bins=bins,color='orange',alpha=0.6,label='Data 2')

plt.xlabel('Very Long Label for X-axis that might overlap with tick labels')
plt.ylabel('Frequency of Occurence (might overlap with bars)')
plt.title('Overlayed Histogram with Overlaping Axis Labels')

plt.legend()

plt.tight_layout()

plt.show()
```

# Nexted pie chart

```python
outer_labels=['Fruits','Vegetables']
outer_sizes=[60,40]
inner_labels=['Apples','Bananas','Carrots','Broccoli']
inner_sizes=[30,30,20,20]
outer_colors=['skyblue','lightgreen']
inner_colors=['lightcoral','orange','yellowgreen','mediumseagreen']
fig,ax=plt.subplots()
ax.pie(outer_sizes,labels=outer_labels,colors=outer_colors,radius=1,wedgeprops=dict(width=0.3,edgecolor='w'))
ax.pie(inner_sizes,labels=inner_labels,colors=inner_colors,radius=0.7,wedgeprops=dict(width=0.3,edgecolor='w'))
plt.title('Nested Pie Chart Example')
plt.show()
```
