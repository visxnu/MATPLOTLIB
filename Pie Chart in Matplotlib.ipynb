# Labels

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylables = ["Apples","Bananas","Cherries","Dates"]

plt.pie(y,labels=mylables)
plt.show()
```

# Startangle

```python
plt.pie(y,labels=mylables,startangle=90)
plt.show()
```

# Explode

```python
myexplode=[0,0,0,0.4]
plt.pie(y,labels=mylables,explode=myexplode)
plt.show()
```

# Shaodow

```python
plt.pie(y,labels=mylables,explode=myexplode,shadow=True)
plt.show()
```

# Color

```python
colorss=['black','red','blue','green']
plt.pie(y,labels=mylables,explode=myexplode,colors=colorss,shadow=True)
plt.show()
```

# Legend

```python
plt.pie(y,labels=mylables,explode=myexplode,colors=colorss,shadow=True)
plt.legend()
plt.show()
```

## Lengend with title

```python
plt.pie(y,labels=mylables,explode=myexplode,colors=colorss,shadow=True)
plt.legend(title='Fruits :')
plt.show()
```
