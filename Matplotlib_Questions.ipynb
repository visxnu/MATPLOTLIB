1. Create a simple line plot with the following y-coordinates: [2, 3, 5, 7, 11].

3. Plot the same data as in question 2, but use square markers ('s') for the data points.

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [2,3,6,8,10]

plt.plot(x,y,marker='s')
plt.show()
```

10. Create a pie chart with the following data: sizes = [20, 30, 40, 10] and labels = ['Red', 'Blue', 'Green', 'Yellow']. Customize the chart with an exploded slice for 'Blue'

```python
sizes = [20, 30, 40, 10]
lab = ['Red', 'Blue', 'Green', 'Yellow']
ex = [0,0.2,0,0]
plt.pie(sizes,labels=lab,explode=ex,shadow=True)
plt.show()
```
